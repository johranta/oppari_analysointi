```python
def fetch_robots_txt(url):
    try:
        response = requests.get(url + '/robots.txt')
        if response.status_code == 200:
            return response.text
        else:
            return f"Failed to fetch robots.txt for {url}. Status code: {response.status_code}"
    except Exception as e:
        return f"An error occurred: {e}"

def is_search_allowed(robots_content):
    lines = robots_content.split('\n')
    disallow_search = False
    for line in lines:
        if line.strip().startswith('Disallow'):
            if 'search' in line.lower() or 'haku' in line.lower():
                disallow_search = True
        elif line.strip().startswith('Allow'):
            if 'search' in line.lower() or 'haku' in line.lower():
                return True
    return not disallow_search

# Lista verkkosivustoista
websites = ['https://www.stockmann.com', 'https://www.k-citymarket.fi', 'https://www.sokos.fi', 'https://www.lumene.com','https://www.lyko.com', 'https://www.tokmanni.fi', 'https://www.prisma.fi', 'https://www.kicks.fi', 'https://www.cocopanda.fi', 'https://www.nordicfeel.fi'  ]

# Taulukko hakemisen sallimisesta tai estämisestä
search_allowance_table = []

# Haetaan ja tutkitaan robots.txt-tiedostot kullekin verkkosivustolle
for site in websites:
    robots_txt_content = fetch_robots_txt(site)
    search_allowed = is_search_allowed(robots_txt_content)
    search_allowance_table.append({'Site': site, 'Search Allowed': search_allowed})

# Tulostetaan taulukko
print("Search Allowance Table:")
for entry in search_allowance_table:
    print(entry)
```
